```python
# importing libraries
import os
import numpy as np
import skimage
from skimage import io
import matplotlib.pyplot as plt
from skimage import data
from skimage.color import rgb2gray
from skimage import filters
from skimage import morphology
from skimage.morphology import skeletonize
from skimage.restoration import (denoise_tv_chambolle, denoise_bilateral,denoise_wavelet, estimate_sigma)
from skimage.filters import threshold_otsu, threshold_local, threshold_minimum
```

```python
# opening lena's picture
lena = io.imread('lena.jpeg')
io.imshow(lena)
io.show()
```

```python
# openin lena's picture the one with noise
lena_noise = rgb2gray(io.imread('lena_noise.jpeg'))
io.imshow(lena_noise)
io.show()
```

```python
# importing c2 librarie for substraction
import cv2
```

```python
# substracting one picture from another
subs_lena = cv2.subtract(lena,lena_noise) 
```

```python
cv2.imshow('Substracted Image', subs_lena)
```

```python
# showing the result 
io.imshow(subs_lena)
io.show()
```
